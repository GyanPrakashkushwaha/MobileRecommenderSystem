# MI Mobiles

## Retrieving features

```python
def bs4_soup(pages):
    URL = f'https://www.flipkart.com/search?q=mi+mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={pages}'
    request = requests.get(URL)
    soup = bs(request.text,'html.parser')
    soup = soup.find('div',class_ = '_1YokD2 _3Mn1Gg')

    return soup
```

```python
def mob_specs(soup):

    names = soup.find_all('div',class_ = '_4rR01T')
    mob_names_list = [i.text for i in names]
    mob_names = [i for i in mob_names_list]


    ratings = soup.find_all('div',class_ = '_3LWZlK')
    mob_ratings = [i.text for i in ratings]

    price = soup.find_all('div',class_ = '_30jeq3 _1_WHN1')
    mob_price = [i.text for i in price]

    img = soup.find_all('div',class_ = 'CXW8mj')
    mob_img_URL = [i.find('img')['src'] for i in img]


    return mob_names,mob_ratings , mob_price , mob_img_URL
    
```

```python
def features_from_mob_links(soup):
    mobile_links = ['https://www.flipkart.com' + i['href'] for i in soup.find_all('a', class_='_1fQZEK')]
    mobile_req = [requests.get(url=i) for i in mobile_links]
    mobile_soup = [bs(i.text, 'html.parser').find('div', class_='_1YokD2 _2GoDe3') for i in mobile_req]
    mobile_specs_soup = [i.find_all('div', class_='_3k-BhJ') for i in mobile_soup]

    storage_ram = [specs_.find('table', class_='_14cfVK').text for mob_specs in mobile_specs_soup for specs_ in mob_specs if 'Memory' in str(specs_.text) and 'Storage' in str(specs_.text)]
    os_processor = [specs_.find('table', class_='_14cfVK').text for mob_specs in mobile_specs_soup for specs_ in mob_specs if 'Os' in str(specs_.text) and 'Processor' in str(specs_.text)]
    camera = [specs_.find('table', class_='_14cfVK').text for mob_specs in mobile_specs_soup for specs_ in mob_specs if 'Camera' in str(specs_.text) and 'FeaturesPrimary' in str(specs_.text)]
    display = [specs_.find('table', class_='_14cfVK').text for mob_specs in mobile_specs_soup for specs_ in mob_specs if 'Display' in str(specs_.text) and 'FeaturesDisplay' in str(specs_.text)]
    network = [specs_.find('table', class_='_14cfVK').text for mob_specs in mobile_specs_soup for specs_ in mob_specs if 'Connectivity' in str(specs_.text) and 'FeaturesNetwork' in str(specs_.text)]
    battery = [specs_.find('table', class_='_14cfVK').text for mob_specs in mobile_specs_soup for specs_ in mob_specs if 'Battery' in str(specs_.text) and 'FeaturesBattery' in str(specs_.text)]

    return mobile_specs_soup , storage_ram, os_processor, camera, display, network, battery
```

```python
mobile_names_list = []
mobile_ratings_list = []
mobile_price_list = []
mobile_image_URL_list = []

mobile_storage_ram_list = []
mobile_camera_list = []
mobile_os_processor_list = []
mobile_display_list = []
mobile_network_list = []
mobile_battery_list = []

page_start = 1

for pages in range(page_start,24):

    soup = bs4_soup(pages=pages)
    mob_names,mob_ratings , mob_price , mob_img_URL = mob_specs(soup)
    mobile_specs_soup , storage_ram, os_processor, camera, display, network, battery = features_from_mob_links(soup=soup)

    if mobile_specs_soup is not None and soup is not None:
        mobile_names_list.append(mob_names)
        mobile_ratings_list.append(mob_ratings)
        mobile_price_list.append(mob_price)
        mobile_image_URL_list.append(mob_img_URL)

        # if counter == 1:

        mobile_storage_ram_list.append(storage_ram)
        mobile_camera_list.append(camera)
        mobile_os_processor_list.append(os_processor)
        mobile_display_list.append(display)
        mobile_network_list.append(network)
        mobile_battery_list.append(battery)

        print(f'scrapping data from page:{pages} and data sample - {mobile_names_list[0] }\n {mobile_ratings_list[0] } \n {mobile_ratings_list[0] } \n specs sample -{mobile_storage_ram_list[0] } \n {mobile_os_processor_list[0] } ')

    else:
        print(f"Skipping page {pages} due to empty mobile_specs_soup or soup.")

```
