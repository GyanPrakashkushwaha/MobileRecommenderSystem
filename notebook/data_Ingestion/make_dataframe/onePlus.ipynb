```python
def make_col(col_name:str):
    col_list_from_df = list(df[col_name])
    col_list = [ast.literal_eval(i) for i in col_list_from_df]
    col_df = [j for i in col_list for j in i]

    return col_df
```

```python
price_df = make_col('price')
```

```python
price_list = []
for i in price_df:
    # print(i)
    if i==None:
        continue
    else:
        price_list.append(i)
```
